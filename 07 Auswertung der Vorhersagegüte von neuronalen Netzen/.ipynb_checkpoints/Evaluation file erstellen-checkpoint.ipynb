```python
xdiss  = []
ydiss  = []
xfor   = []
label  = []
for j in range(50):
    xdiss.append(locals()["xdisplacement1"+str(j)]) 
    ydiss.append(locals()["ydisplacement1"+str(j)])
    xfor.append(locals()["xforc1"+str(j)])
    label.append(locals()["labell1"+str(j)])

xdiss = np.array(xdiss)
ydiss = np.array(ydiss)
xfor = np.array(xfor)
label  = np.array(label)
xdis = xdiss.transpose(0,2,1)
ydis = ydiss.transpose(0,2,1)

####################################################################################################################
print(xdis.shape)
print(ydis.shape)
print(xfor.shape)
print(label.shape)
```

```python
xyfornotscaled = np.stack((xdis,ydis,xfor),3)
#xyfor1 = np.stack((xdisscaled,ydisscaled),3)
print(xyfornotscaled.shape)
```

```python
from sklearn.metrics import mean_squared_error                           # mean_squared_error berechnen

A = np.arange(10000,10050)
B = np.arange(10300,10350)
MSE = []
for i , j in zip(A, B):
    mse = mean_squared_error(xyfornotscaled[i,132,0:43,2], xyfornotscaled[j,132,0:43,2])
    MSE.append(mse)

MSE = np.array(MSE)
MSE = np.sort(MSE)
print(MSE.shape)
print(MSE)
```

```python
#Plotting filtering
plt.figure( figsize=(14,7))    # In this example, all the plots will be in one figure. 
plt.title('MSE Distribution for validation sets ',family='serif', style='italic' ,fontsize=18)

x = np.arange(50)
y = MSE

plt.bar(x, y,width=0.8,color='r',label='MSE')
plt.xlabel('Validation set number',family='serif', style='italic' ,fontsize='x-large')
plt.ylabel('Mean Squared Error(MSE)',family='serif', style='italic' ,fontsize='x-large')
plt.grid( ls=':', color='#AAAAAA', linewidth=1.8 )
plt.legend(edgecolor='blue' , fontsize='large')
#print(yfilter)
plt.savefig('/content/drive/MyDrive/Vergleich/MSE.png')
```
