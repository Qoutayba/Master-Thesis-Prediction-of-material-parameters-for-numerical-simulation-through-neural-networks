```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import matplotlib.pyplot as plt
```

```python
h5file = h5py.File('/content/drive/MyDrive/Finally training Data/Finally Training data 10150 not scaled.h5', 'r' )
```

```python
xyfornotscaled = np.array(h5file['Training data not scaled/xyfornotscaled'])
label = np.array(h5file['Training data not scaled/Labels'])
print(xyfornotscaled.shape)
print(label.shape)
h5file.close()
```

```python
XYFORnotscaled = xyfornotscaled.reshape((xyfornotscaled.shape[0], 272*101*3))
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(XYFORnotscaled)
xyforscaled = scaler.transform(XYFORnotscaled)
```

```python
xyfor = xyforscaled.reshape((xyforscaled.shape[0], 272,101,3))
print(xyfor.shape)
```

```python
imag1  = xyfor[4,:,:,:]
imag2  = xyfor[6,:,:,:]
#imag0    = xdisscaled[0]
#imag1    = xforscaled[0]
#imag10    = xforscaled[10]
print(imag1.shape)
print(imag1)
print(imag2)
```

```python
fig, axes = plt.subplots( 1,2,figsize=(16,10))
axes[0].imshow( imag1  )
axes[1].imshow( imag2 )
axes[0].set_title('Image1')
axes[1].set_title('Image2')
```

```python
filename = '/content/drive/MyDrive/Finally training Data/Finally Training data 10150 Standardscaler.h5'
h5file = h5py.File( filename, 'w' )
h5file.create_group('Training data')
h5file['Training data'].create_dataset('xyfor', data = xyfor)
h5file['Training data'].create_dataset('Labels', data = label)   
```
