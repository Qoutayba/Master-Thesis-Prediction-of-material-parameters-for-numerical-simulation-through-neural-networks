```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import h5py
from lasso.dyna import Binout
from math import e
import sys
import os
```

## Create an array of all cases

```python
VarNum=6                          # number of Variable.
SPoints=10                        #number of support Pionts.
n= SPoints**VarNum                # number of Cases
m=VarNum

MinE=2.1                          # Minimum value for E Modul
MaxE=2.6                          # Maximum value for E Modul
MinPR=0.36                        # Minimum value for PR Poisson factor
MaxPR=0.41                        # Maximum value for PR Poisson factor
MinAA=0.1                         # Minimum value of A (The first variable in the Hockett-Shorby equation)
MaxAA=0.2                         # Maximum value of A (The first variable in the Hockett-Shorby equation)
MinBB=0.01                        # Minimum value of B (The second variable in the Hockett-Shorby equation)
MaxBB=0.09                        # Maximum value of B (The second variable in the Hockett-Shorby equation)
MinCC=0.1                         # Minimum value of C (The third variable in the Hockett-Shorby equation)
MaxCC=1.1                         # Maximum value of C (The third variable in the Hockett-Shorby equation)
MinNN=0.5                         # Minimum value of N (The fourth variable in the Hockett-Shorby equation)
MaxNN=1.1                         # Maximum value of N (The fourth variable in the Hockett-Shorby equation)

E =np.arange(MinE ,MaxE+0.00001,(MaxE-MinE)/(SPoints-1))
E=np.around(E, decimals=2)
PR=np.arange(MinPR,MaxPR+0.00001,(MaxPR-MinPR)/(SPoints-1))
PR=np.around(PR, decimals=3)
AA=np.arange(MinAA,MaxAA+0.00001,(MaxAA-MinAA)/(SPoints-1))
AA=np.around(AA, decimals=2)
BB=np.arange(MinBB,MaxBB+0.00001,(MaxBB-MinBB)/(SPoints-1))
BB=np.around(BB, decimals=2)
CC=np.arange(MinCC,MaxCC+0.00001,(MaxCC-MinCC)/(SPoints-1))
CC=np.around(CC, decimals=2)
NN=np.arange(MinNN,MaxNN+0.00001,(MaxNN-MinNN)/(SPoints-1))
NN=np.around(NN, decimals=2)
print('E    = ',E )
print('PR   = ',PR)
print('AA   = ',AA)
print('BB   = ',BB)
print('CC   = ',CC)
print('NN   = ',NN)
print('number of Cases is    ',n)
Allcases = []
for k in E:
   
    for i in PR:
   
       for j in AA:
           
             for l in BB:
                    
                    for q in CC:
                        
                        for z in NN:
        
                              onecase = [k,i,j,l,q,z]      # Only one csae in each Loop
                              Allcases.append(onecase)     # All Casees in one Matrix 

print('Shape of Matrix of all cases   ',np.shape(Allcases))
print(" ")
print('All cases of (E   PR   A   B   C   N):')
#print(Allcases)
```

## Reshape and shuffle the matrix then choose a specific number of cases from the matrix of all cases

```python
##################################################  Reshape  #######################################
allcases = np.reshape(Allcases,(n,m))
##################################################  shuffle  #######################################
np.random.shuffle(allcases)
####################################################################################################
#print("allcases = ", allcases)
print(np.shape(allcases))
```

```python
##################################  select a specific number of allcases  ##########################
idx = np.arange(10100)                             # here take anumber of cases but not all (in our case we need 10000 cases für Training and 100 cases für validation)
A   = allcases[idx,:] 
#print(idx)
print('shape of A = ',A.shape)
print('number of cases = ', A.shape[0])
print(A)
```

## Write the cases in the Keyfile Cases.k

```python
Ca = open("Cases2100.k","w")
for i in range(10100):
    Ca.write('*CASE_BEGIN_{}\n'.format(i))
    Ca.write('*PARAMETER\n')
    Ca.write('$#   prmr1      val1     prmr2      val2     prmr3      val3     prmr4      val4\n')
    Ca.write('RE         {}      RPR        {}\n'.format(A[i,0],A[i,1]))
    Ca.write('RA         {}       RB         {}       RC         {}       RN         {}\n'.format(A[i,2],A[i,3],A[i,4],A[i,5]))
    Ca.write('*CASE_END_{}\n'.format(i))
Ca.write('*INCLUDE\n')
Ca.write('Mat024withoutFailure.k\n')
Ca.write('*INCLUDE\n')
Ca.write('Probe117b1F-Implicit.k\n')
Ca.write('*END')
Ca.close()
```

## Write the cases as labels in HDF5 file 

```python
filename = 'Training data 10100 only Labels.h5'
h5file = h5py.File( filename, 'w' )
for i in range (A.shape[0]):
    b = 'case{}'.format(i)
    h5file.create_group(b)
    h5file[b].create_dataset('Label{}'.format(i), data = A[i,:])
    h5file[b].attrs.update(dict(E = A[i,0] , Pr = A[i,1] , A = A[i,2] , B = A[i,3] , C = A[i,4] , N = A[i,5]))
#h5file.visit(print)
#print( 'metadata cases:\n', dict( h5file[b].attrs) )
h5file.close()
```

### Plotting the curves of Hocket sherby and compare them with Hardening curve of PC

```python
##############################################   Hardening curve   #################################################

Hardeningcurve = pd.read_csv( 'data/Hardening-curve.csv', sep='\s+' , engine='python' )
Hardeningcurve.head(101)
Hardeningcurve.describe()
XHardening = Hardeningcurve['xa'].unique()
YHardening = Hardeningcurve['ya'].unique()
#print('X-Value = ', XHardening , type(XHardening))
#print('Y-Value = ', YHardening , type(YHardening))

###################   Ploting Hardening curve with the curves of all cases or some of them   #######################

fig, ax = plt.subplots(figsize=(12,7))
for i in range(500):
    Xhockettsherby = np.arange(0.0,3.02,0.01)
    yhockettsherby = A[i,2]-(A[i,3]*(e**(-A[i,4]*Xhockettsherby**(A[i,5]))))
    ax.plot(Xhockettsherby, yhockettsherby, lw=1.5, c='r')
ax.plot( XHardening, YHardening, label=r'Hardening curve',lw=2.5 )
firsthockettsherby = A[0,2]-(A[0,3]*(e**(-A[0,4]*Xhockettsherby**(A[0,5]))))
ax.plot(Xhockettsherby, firsthockettsherby, lw=1.5, c='r' , label='Hockett sherby cases')
ax.set_xlim( XHardening.min(), XHardening.max() )
ax.set_title('one Hardening curve with different Hockett sherby cases',family='serif', style='italic' ,fontsize='x-large')
ax.grid( ls='--', color='gray', linewidth=0.5 )
ax.legend(edgecolor='blue' , fontsize='large')
ax.set_ylabel('Hockett sherby (x) [-]' , family='serif', style='italic' , fontsize='x-large')
ax.set_xlabel('X [-]' , family='serif', style='italic' , fontsize='x-large')
plt.show()
```
